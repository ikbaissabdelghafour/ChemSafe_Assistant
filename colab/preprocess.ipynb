# Preprocessing Notebook

This notebook demonstrates the pipeline to clean, parse, convert SMILES to molecular graphs, and split the Tox21 dataset.

```python
import os
import sys

# Ensure project root is in system path for imports
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

import torch
from src.data.dataset import save_processed_pipeline, TOX21_LABELS
```

```python
# Define file paths
csv_path = "../data/raw/tox21.csv"
save_path = "../data/processed/graphs.pt"

# Run preprocessing and splitting pipeline
data_splits = save_processed_pipeline(csv_path, save_path, desalt=True, seed=42)
```

```python
# Inspect a processed sample graph
train_set = data_splits['train']
if len(train_set) > 0:
    sample_graph = train_set[0]
    print("--- Sample Graph Info ---")
    print(f"SMILES: {sample_graph.smiles}")
    print(f"Number of Atoms (Nodes): {sample_graph.num_nodes}")
    print(f"Atom Feature Matrix Shape: {sample_graph.x.shape}")
    print(f"Number of Bonds (Edges): {sample_graph.num_edges}")
    print(f"Edge Index Shape: {sample_graph.edge_index.shape}")
    print(f"Edge Feature Matrix Shape: {sample_graph.edge_attr.shape}")
    print(f"Labels Tensor: {sample_graph.y}")
    
    # Print class label map
    for name, val in zip(TOX21_LABELS, sample_graph.y.tolist()):
        print(f"  {name}: {val}")
else:
    print("No graphs in train set. Check that data/raw/tox21.csv contains valid data.")
```
